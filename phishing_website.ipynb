```python
# import libs
from urllib.parse import urlparse,urlencode
import ipaddress
import re
```

```python
def getDomain(url):
  domain = urlparse(url).netloc
  if re.match(r"^www.", domain):
    domain = domain.replace("www.", "")
  return domain
```

```python
#show just url column
dataPhishingUrl = urlPhishing[['url']]
dataPhishingUrl.head()
```

```python
urls = dataPhishingUrl.url.values
urls
```

```python
domainNames = [0]*len(urls)
for i in range(0, len(urls)):
  domainNames[i] = getDomain(urls[i])
len(domainNames)
```
